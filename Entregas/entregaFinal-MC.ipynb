# Phishing URL Detection

El dataset elegido es Phishing URL Detection de la web Kaggle.com.
https://www.kaggle.com/datasets/sergioagudelo/phishing-url-detection?resource=download

Este dataset sobre detección de phishing es una colección extensa de URLs, en los que se encuentra la variable target elegida, la cual figura como "legítima" o de "phishing".
Esta permitiría facilitar la investigación y desarrollo de algoritmos de clasificación para identificar sitios web fraudulentos. Esta base de datos, con aproximadamente 2.5 millones de muestras, abarca URLs etiquetadas como phishing o legítimas, obtenidas de repositorios principales de URLs de phishing (como PhishTank y Phishing.Database) y sitios de clasificación de URLs (como Cisco Umbrella Popularity List y The Majestic Million). La finalidad del creador de este dataset es abordar la falta de grandes conjuntos de datos etiquetados y/o herramientas de sandboxing que permitan una importación sencilla de muestras de URL para la detección de phishing.



## 1. Definición de Objetivo
El objetivo de este análisis es desarrollar un modelo de clasificación que permita detectar si una URL es legítima o corresponde a un ataque de phishing, utilizando un conjunto de 15 características relevantes.



## 2. Contexto Comercial
El phishing es una táctica común utilizada por atacantes para engañar a los usuarios, haciéndoles creer que están visitando un sitio legítimo cuando en realidad es fraudulento. El costo de estos ataques es alto, afectando tanto a las empresas como a los usuarios individuales.



## 3. Problema Comercial
La necesidad de identificar y prevenir ataques de phishing es crítica para reducir el fraude online y mantener la confianza de los usuarios en las plataformas de internet. Un sistema automatizado que detecte URLs maliciosas puede mejorar significativamente la seguridad online.



## 4. Contexto Analítico
Utilizaremos el dataset seleccionado con las características de URLs para construir un modelo de clasificación. El análisis inicial (EDA) permitirá comprender las relaciones entre las características y la etiqueta de phishing, lo que servirá de base para la selección de características y el algoritmo a utilizar.


El dataset incluye 15 características relevantes para la detección de phishing, que se describen a continuación:

**url**: La cadena de caracteres que representa la URL.

**source**: La fuente de la URL, ya sea de URLs de phishing o de bases de datos de clasificación de URLs.

**label**: La categoría de la URL, que puede ser phishing o legítima.

**url_length**: La longitud de la URL en caracteres.

**starts_with_ip**: Indica si la URL base es una dirección IP.

**url_entropy**: La entropía de la URL o el nombre de host, una medida de la aleatoriedad en la URL.

**has_punycode**: Indica si la URL contiene al menos un carácter punycode, utilizado para representar caracteres no ASCII.

**digit_letter_ratio**: La relación entre dígitos y letras en la URL.

**dot_count**: El número de ocurrencias del punto (.) dentro de la URL.

**at_count**: El número de ocurrencias del símbolo (@) dentro de la URL.

**dash_count**: El número de ocurrencias del guion (-) dentro de la URL.

**tld_count**: El número de dominios de nivel superior en el subdirectorio de la URL.

**domain_has_digits**: Indica si el dominio base de la URL contiene dígitos.

**subdomain_count**: El número de subdominios presentes en la URL base.

**nan_char_entropy**: La entropía de los caracteres no alfanuméricos dentro de la URL.

**has_internal_links**: Indica si el subdirectorio de la URL contiene enlaces internos.

**whois_data**: Datos del registro WHOIS del dominio.

**domain_age_days**: La edad del dominio en días, según el registro WHOIS extraído.

Este dataset podría proporcionar una base de datos para construir modelos de clasificación de phishing, permitiendo a los investigadores y desarrolladores evaluar y mejorar la precisión de sus sistemas de detección mediante el análisis de diversas características de las URLs.

## 6. Data Engineering - Selección de Características
Para reducir la dimensionalidad del dataset, aplicaremos un método de selección de características basado en la importancia calculada a partir de modelos de árboles de decisión.


### 6. 1. Dividir los Datos
Dividimos el dataset en conjunto de entrenamiento y prueba.

Eliminamos la columna 'label' para usarla como objetivo, en consecuencia label_num, y url por ser no numérica e irrelevante.

#### Dado que nan_char_entropy y url_entropy estan relacionadas con label de manera muy directa también las eliminamos.

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=['label','url','whois_data','label_num','source','nan_char_entropy','url_entropy']) 
y = df['label'].map({'phishing': 1, 'legitimate': 0})  # Mapeamos las etiquetas a valores binarios

# Divido en conjunto de entrenamiento y prueba (70% entrenamiento, 30% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

## 7. Selección del Algoritmo Adecuado
Utilizaremos un algoritmo de clasificación como **Random Forest**, ya que son efectivos para este tipo de problemas de clasificación. Este modelo también permite manejar automáticamente características irrelevantes.


### 7. 1. Selección de Características
Usamos tambipén Random Forest para hacer Feature Selection y elegir las características más importantes:

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

importances = rf.feature_importances_
feature_importance = sorted(zip(importances, X.columns), reverse=True)

# se muestran las características más importantes
for importance, feature in feature_importance[:10]:
    print(f"Feature: {feature}, Importance: {importance}")
```

## 8. Desarrollo del Algoritmo
Entrenaremos el modelo utilizando el conjunto de entrenamiento creado, y lo evaluaremos utilizando las siguientes métricas: exactitud, precisión, sensibilidad, especificidad y F1 score.


### 8. 1. Entrenamiento del Modelo con las Características Seleccionadas

Elegimos las 10 características más importantes y entrenamos el modelo:

```python
# se seleccionan las 10 características más importantes
important_features = [feature for importance, feature in feature_importance[:10]]

# filtro las características seleccionadas en los conjuntos de entrenamiento y prueba
X_train_selected = X_train[important_features]
X_test_selected = X_test[important_features]

# acá se entrena el modelo
rf_selected = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,           # Limita la profundidad de los árboles
    min_samples_split=5,    # Mínimo de muestras para dividir un nodo
    min_samples_leaf=2,     # Mínimo de muestras en hojas
    max_features='sqrt',    # Limita características consideradas en cada split
    random_state=42
)
rf_selected.fit(X_train_selected, y_train)
```

### 8. 2. Evaluamos el Modelo
Calculamos las métricas para evaluar el desempeño del modelo:

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# se hacen las predicciones en el conjunto de prueba
y_pred = rf_selected.predict(X_test_selected)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# se hace la matriz de confusión para obtener la especificidad
conf_matrix = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = conf_matrix.ravel()
specificity = tn / (tn + fp)

# Mostrar resultados
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"Specificity: {specificity}")
print(f"F1 Score: {f1}")
```
